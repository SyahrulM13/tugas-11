```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
n = 1000

data = {
    "CustomerID": np.arange(1, n+1),
    "Age": np.random.randint(18, 65, size=n),
    "Gender": np.random.choice(["Male", "Female"], size=n),
    "AnnualIncome": np.round(np.random.normal(100, 30, size=n), 2),
    "SpendingScore": np.random.randint(1, 101, size=n),
    "MembershipLevel": np.random.choice(["Silver", "Gold", "Platinum"], size=n, p=[0.5, 0.3, 0.2]),
    "City": np.random.choice(["Jakarta", "Bandung", "Surabaya", "Yogyakarta", "Medan"], size=n),
    "TimeOnApp": np.round(np.random.normal(2.5, 1.0, size=n), 2),
    "LastPurchaseDays": np.random.randint(0, 90, size=n),
    "Target": np.random.choice([0, 1], size=n, p=[0.6, 0.4])
}

df = pd.DataFrame(data)

print("Contoh data:")
print(df.head())

print("\nStatistik deskriptif:")
print(df.describe(include='all'))

print("\nDistribusi target:")
print(df['Target'].value_counts(normalize=True))

print("\nKorelasi antar variabel numerik:")
print(df.corr(numeric_only=True))
```
